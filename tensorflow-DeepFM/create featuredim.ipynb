```python
import pandas as pd
import pickle
```

```python
dfTrain = pd.read_csv('robust_train.csv')
dfTest = pd.read_csv('robust_test.csv')
```

```python
df = pd.concat([dfTrain, dfTest])
```

```python
CATEGORICAL_COLS = [
    "cate_id", "campaign_id", "customer",
    "brand", "pid", "final_gender_code",
    "age_level", "pvalue_level", "shopping_level",
    "occupation", "city"
]

NUMERIC_COLS = [
    "price", "cart", "fav",
    "pv", "buy"
]

IGNORE_COLS = [
    "user", "adgroup_id", "clk",
    "nonclk", "cms_segid", "cms_group_id"
]

feat_dict = {}
tc = 0
for col in df.columns:
    if col in IGNORE_COLS:
        continue
    if col in NUMERIC_COLS:
        # map to a single index
        feat_dict[col] = tc
        tc += 1
    else:
        us = df[col].unique()
        feat_dict[col] = dict(zip(us, range(tc, len(us)+tc)))
        tc += len(us)
```

```python
feat_dim = tc
```

```python
feat_dict['dim'] = 781089
```

```python
with open("feat_dict.pkl", "wb") as f:
    pickle.dump(feat_dict, f, pickle.HIGHEST_PROTOCOL)
```
